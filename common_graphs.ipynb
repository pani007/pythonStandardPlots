```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
%matplotlib inline
```

## Lets plot `sin` , `exp` and `linear`  and `circle`

the below function impliments all plots

```python

## x_arr is an iterator for x axis any sort or array can be used
## for sinusoid make sure x axis is in radians
## only need the x axis m and c are only used when you select linear otherwise its optional parameter
## Implimented circle in ths function will impliment other urves in future




def plot(x_arr,category,slope = 1,intercept =0): 
    
    if(category == 'sin'):
        y = [round(m.sin(i),2) for i in x_arr]
        plt.plot(x_arr,y)
        plt.xlabel('x')
        plt.ylabel('y')
        
    elif category == 'linear':
        y = [slope*i + intercept for i in x_arr]
        plt.plot(x_arr,y, color='green')
        plt.xlabel('x')
        plt.ylabel('y')
        
    elif category == 'exp':
        y = [round(m.exp(i),2) for i in x_arr]
        plt.plot(x_arr,y, color='red')
        plt.xlabel('x')
        plt.ylabel('y')
        
    elif category == 'circle':
        r = round(max(x_arr),0)
        y = [(r**2 - i**2)**0.5 for i in x_arr]
        plt.plot(x_arr,y,color='blue')
        y_inv = [-(r**2 - i**2)**0.5 for i in x_arr]
        plt.plot(x_arr,y_inv,color='blue')
    
    else: 
        print('invalid category')
```

### For sin wave your x axis should be radians

```python
x_sin = [i* m.pi/180 for i in range(720)] ## 360 is one period 2* pi and 720 is 4*pi hence 2 periods
plot(x_sin,'sin')
plt.ylim([-1, 1])
plt.xlim([0, 15])
```

   ### Lets plot  linear
   - slope = 1.73 inclination is 60 degrees  
   - intercept 3

```python
x_lin = list(range(0,10))
plot(x_lin,'linear',3**0.5,3)
plt.ylim([-3, 30])
plt.xlim([0, 15])
```

## make sure the x range is between `-radius` to `+radius` 
## x array should  atleast have 100 samples
### more the samples more smooth circle will be
###  circle will have center at origin

```python
plt.figure(figsize=(7,7))## x and y width should be same otherwise it will look like elipse
x_cir = list(np.arange(-3,3.1,0.006)) 
plot(x_cir,'circle')
```
